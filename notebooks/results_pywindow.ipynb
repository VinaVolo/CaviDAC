```python
import os
import sys
sys.path.append("..")

import pandas as pd
import pywindow as pw

from CalixVolApp.utils.paths import get_project_path
```

# Calculate Volume for all molecules

```python
molecules, volumes = [], []

for mo in range(14):
    
    molec = pw.MolecularSystem.load_file(os.path.join(get_project_path(), 'CalixVolApp', 'data', 'molecules', 'pdb_calix', f'{mo+1}.pdb'))
    mol = molec.system_to_molecule()
    
    volume_cavity = mol.calculate_pore_volume_opt()
    
    volumes.append(volume_cavity / 2 if mol == 14 else volume_cavity)
    
    molecules.append(mo+1)
    
df_res = pd.DataFrame({'Molecule': molecules, 'Cavity_Volume': volumes})
df_res.to_csv(os.path.join(get_project_path(), 'results', 'pywindow', 'calc_volumes_pywindow.csv'), index=False)

df_res
```
